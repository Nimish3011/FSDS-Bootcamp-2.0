**Problem Statment : Find out every IPL season winner**

```python
data.drop_duplicates(subset=['city','season']).shape
```

```python
data.drop_duplicates('season') # By default, the function takes the first value and returns the output. The second time it is encountered, it is skipped because it is considered a duplicate.
```

```python
data.drop_duplicates('season',keep='last')[['season','winner']].sort_values('season') # keep = "last" is  keeping only the last occurrence of each season. This means that if there are multiple rows with the same season value, only the one with the latest occurrence in the DataFrame will be kept. The resulting DataFrame will have a unique list of seasons.
```
